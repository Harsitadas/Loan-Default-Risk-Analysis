7. Filter the ‘Emp_Len’ column to extract the numerical value from the string Hint - Emp_len : < 1year, 2 years , 3 years as 1 , 2, 3 so on.


```python
def extract_numerical(emp_length):
    if pd.isnull(emp_length):
        return None
    elif emp_length.startswith('<'):
        return 0
    elif emp_length.startswith('10+'):
        return 10
    else:
        return int(emp_length.split()[0])
    
```

```python
 df['emp_length']=df['emp_length'].apply(extract_numerical)
```

11.Using a user defined function convert the ‘emp_len’ column into
categorical column as follows -
If emp_len is less than equals to 1 then recode as ‘fresher’.
If emp_len is greater than 1 and less than 3 then recode as ‘junior’.
If emp_len is greater than 3 and less than 7 then recode as ‘senior’
If emp_len is greater than 7 then recode as ‘expert’.

```python
def categorize_emp_len(emp_length):
    if emp_length <= 1:
        return 'fresher'
    elif 1 < emp_length < 3:
        return 'junior'
    elif 3 <= emp_length < 7:
        return 'senior'
    elif emp_length >= 7:
        return 'expert'
    else:
        return 'unknown'
    
df['emp_category'] = df['emp_length'].apply(categorize_emp_len)
df
```

12.Find the sum of ‘loan_amnt’ for each grade and display the distribution
of ‘loan_amnt’ using a pie plot.

```python
sum_by_grade = df.groupby('grade')['loan_amnt'].sum()
plt.figure(figsize=(8, 8))
plt.pie(sum_by_grade, labels=sum_by_grade.index, autopct='%1.1f%%',startangle=150)
plt.title('Distribution of loan_amnt by Grade')
plt.show()
```
